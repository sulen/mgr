```python
import matplotlib.pyplot as plt

# Data points
methods = ['Bag of Words', 'TF-IDF', 'Word2Vec', 'GloVe', 'Ember-v1']
accuracy = [0.81, 0.79, 0.59, 0.67, 0.69]

# Create bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(methods, accuracy)

# Adding title and labels
plt.xlabel('Methods')
plt.ylabel('F1-Score')
# for bar in bars:
#     yval = bar.get_height()
#     plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

plt.savefig('accuracy_chart.png')
# Display the chart
plt.show()
```
